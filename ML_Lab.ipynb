```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.ensemble import RandomForestClassifier

# Load the dataframe
try:
    df = pd.read_csv('titanic_train (1).csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: 'titanic_train (1).csv' not found. Please upload the file.")
    # Exit the cell execution if the file is not found
    raise

# Define features (X) and target (y)
X = df.drop(labels=['Survived', 'PassengerId'], axis=1)
y = df['Survived']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=0)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")

# Identify categorical and numerical features
categorical_features = X_train.select_dtypes(include=['object']).columns
numerical_features = X_train.select_dtypes(include=np.number).columns

# Create preprocessing pipelines for numerical and categorical features
numerical_transformer = SimpleImputer(strategy='median') # Handle missing numerical values
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')), # Handle missing categorical values
    ('onehot', OneHotEncoder(handle_unknown='ignore')) # Apply one-hot encoding
])

# Create a column transformer to apply different transformations to different columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough' # Keep other columns (if any) - though not expected here
)

# Create a pipeline that first preprocesses the data and then applies the classifier
# We use a dummy classifier here just for SFS, the actual model will be inside SFS
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Apply preprocessing to training data to get the processed feature names
X_train_processed = pipeline.fit_transform(X_train)

# Get the feature names after preprocessing
# This part is a bit tricky with ColumnTransformer and OneHotEncoder
# One way is to manually construct or inspect after fitting the preprocessor
# For simplicity, we will use the original column names for SFS input
# and rely on SFS to work with the transformed numpy array.
# If you need the exact names after one-hot, you would need more complex code.
# For SFS, passing the numpy array from the preprocessor is standard.

# Initialize Sequential Feature Selector
sfs1 = SFS(RandomForestClassifier(n_jobs=4, random_state=0), # Added random_state for reproducibility
           k_features=5,   # select the best 5 features
           forward=True,  # adds one feature at a time
           verbose=2,  # progress printed in the console
           scoring='roc_auc',
           cv=3) # Use cross-validation

# Fit SFS on the preprocessed training data
# We need to apply the full preprocessing pipeline before fitting SFS
# Let's create a pipeline that includes preprocessing and the classifier for SFS
# Note: SFS fits the estimator internally, so we just need the preprocessed data
# and the base estimator for SFS itself.
# We will fit the preprocessor separately and then use the transformed data for SFS.

# Fit SFS on the preprocessed training data (handled missing values and encoding)
# Use the preprocessor to transform X_train before passing to SFS
X_train_transformed = preprocessor.fit_transform(X_train)

# Now fit SFS using the transformed data
sfs1 = sfs1.fit(X_train_transformed, y_train)


# Print the results
print('\nSequential Feature Selection Results:')
print('Best accuracy score: %.2f' % sfs1.k_score_)   # k_score_ shows the best score
# The k_feature_idx_ and k_feature_names_ refer to the indices/names
# of the features in the *input* array to SFS.
# If input is a numpy array from preprocessor, indices are array indices.
# Getting original feature names from indices of transformed array is complex.
# We will just print the indices from the transformed array.
print('Best subset (indices in transformed array):', sfs1.k_feature_idx_)
# To get original feature names, you would need to map back from the indices
# of the one-hot encoded array to the original features, which is non-trivial
# with ColumnTransformer and handle_unknown='ignore'.
# For now, we focus on getting SFS to run.
```

```python
#Prepare the Data (pca)
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming your DataFrame is named 'df'
df = pd.read_csv('house_price_train (1).csv')

# Select numerical features for PCA
# These are chosen as an example. You can select more or different features.
features = ['OverallQual', 'GrLivArea', 'GarageArea', 'TotalBsmtSF', 'YearBuilt']
X = df[features]

# Handle any potential missing values by dropping rows with NaN
X = X.dropna()

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Apply PCA
from sklearn.decomposition import PCA

# Apply PCA with all components to understand variance
pca = PCA()
pca.fit(X_scaled)

# Print the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
print("Explained Variance Ratio:")
print(explained_variance_ratio)

# Sum to see total variance captured by the first few components
cumulative_variance = pca.explained_variance_ratio_.cumsum()
print("\nCumulative Explained Variance:")
print(cumulative_variance)
```

```python
#Transform Data and Visualize
import matplotlib.pyplot as plt

# Re-run PCA with the desired number of components (e.g., 2)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plot the principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df.loc[X.index, 'SalePrice'], cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('2D PCA of House Features')
plt.colorbar(label='SalePrice')
plt.show()
```

```python
#Using PCA Output for Logistic Regression
# Assuming you have already performed the PCA steps and have X_pca and y
# (where y is your target variable, e.g., 'is_expensive')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the new PCA-transformed data
X_pca_train, X_pca_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Initialize and train the Logistic Regression model on the PCA data
model_pca = LogisticRegression()
model_pca.fit(X_pca_train, y_train)

# Make predictions and evaluate the model
y_pred_pca = model_pca.predict(X_pca_test)
accuracy_pca = accuracy_score(y_test, y_pred_pca)

print(f'Accuracy of the Logistic Regression model with PCA: {accuracy_pca:.2f}')
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Load the dataset
df = pd.read_csv('house_price_train (1).csv')

# Create a binary target variable
median_price = df['SalePrice'].median()
df['is_expensive'] = (df['SalePrice'] > median_price).astype(int)

# Select features and the new target
features = ['OverallQual', 'GrLivArea']
X = df[features]
y = df['is_expensive']

# Align y with X after dropping NaNs
X = X.dropna()
y = y.loc[X.index]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create a numerical version of y for plotting
y_numeric = pd.factorize(y)[0]

# Create the side-by-side plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot Before PCA
sc1 = ax1.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_numeric, cmap='coolwarm', edgecolor='k', s=80)
ax1.set_xlabel('Standardized OverallQual')
ax1.set_ylabel('Standardized GrLivArea')
ax1.set_title('Before PCA: Using Original Standardized Features')
cbar1 = fig.colorbar(sc1, ax=ax1, ticks=range(len(pd.factorize(y)[1])))
cbar1.set_label('Target classes')
cbar1.set_ticklabels(['Not Expensive', 'Expensive'])

# Plot After PCA
sc2 = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=y_numeric, cmap='coolwarm', edgecolor='k', s=80)
ax2.set_xlabel('Principal Component 1')
ax2.set_ylabel('Principal Component 2')
ax2.set_title('After PCA: Projected onto 2 Principal Components')
cbar2 = fig.colorbar(sc2, ax=ax2, ticks=range(len(pd.factorize(y)[1])))
cbar2.set_label('Target classes')
cbar2.set_ticklabels(['Not Expensive', 'Expensive'])

plt.tight_layout()
plt.show()
```
